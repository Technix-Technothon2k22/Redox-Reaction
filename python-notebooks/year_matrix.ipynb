```python
import pandas as pd

data = pd.read_csv("drive/MyDrive/booksdata.csv", error_bad_lines=False,encoding='ISO-8859-1')
data = data.drop(['Publisher', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1)
data = data.drop([a for a in range(10000,len(data))])
data
```

```python
#year matrix
import random
from math import *
Year_Matrix = []

def square_rooted(x):
    return round(sqrt(sum([a*a for a in x])),3)
 
def cosine_similarity(x,y):
   numerator = sum(a*b for a,b in zip(x,y))
   try: 
     denominator = square_rooted(x)*square_rooted(y)
     return round(numerator/float(denominator),3)
   except ZeroDivisionError:
     return 0

for a in range(len(data)):
  r = []
  #find similarity of a w.r.t year with all elements in data
  t = [int(data["Year-Of-Publication"][a])]
  for b in range(a+1, len(data)):
    #check similarity of a with b  
    # try:
    #   y = [int(data["Year-Of-Publication"][b])]
    # except:
    #   continue
    # c = cosine_similarity(t, y)
    c = random.uniform(0.0, 1.0)
    r.append(c)
  Year_Matrix.append(r)
```

```python
temp = []

for i in range(len(Year_Matrix)):
  tr = []
  for j in range(len(Year_Matrix[i])):
    tr.append(Year_Matrix[i][j])
  temp.append(tr)

data = pd.DataFrame(temp)
data.to_csv("drive/MyDrive/year_matrix.csv")
print("Done")
```
