```python
import pandas as pd

title_data = pd.read_csv("drive/MyDrive/title_matrix.csv", error_bad_lines=False,encoding='ISO-8859-1')
author_data = pd.read_csv("drive/MyDrive/author_matrix.csv", error_bad_lines=False,encoding='ISO-8859-1')
year_data = pd.read_csv("drive/MyDrive/year_matrix.csv", error_bad_lines=False,encoding='ISO-8859-1')
rating_data = pd.read_csv("drive/MyDrive/rating_matrix.csv", error_bad_lines=False,encoding='ISO-8859-1')
age_data = pd.read_csv("drive/MyDrive/age_matrix.csv", error_bad_lines=False,encoding='ISO-8859-1')
```

```python
print(f"Title\trows: {title_data.shape[0]}\tcolumns: {title_data.shape[1]}")
print(f"Author\trows: {author_data.shape[0]}\tcolumns: {author_data.shape[1]}")
print(f"Year\trows: {year_data.shape[0]}\tcolumns: {year_data.shape[1]}")
print(f"Rating\trows: {rating_data.shape[0]}\tcolumns: {rating_data.shape[1]}")
print(f"Age\trows: {age_data.shape[0]}\tcolumns: {age_data.shape[1]}")
```

```python

for a in range(title_data.shape[0]):
  #find similarity of a w.r.t all elements in data
  for b in range(a, title_data.shape[1]):
    #check similarity of a with b
    print(title_data.iat[a,b])
```

```python
Main_Matrix = []

for a in range(title_data.shape[0]):
  r = []
  #find similarity of a w.r.t all elements in data
  for b in range(a, title_data.shape[1]):
    #check similarity of a with b
    weighted_avg = (title_data.iat[a,b] + author_data.iat[a,b] + year_data.iat[a,b] + rating_data.iat[a,b] + age_data.iat[a,b]) / 5
    r.append(weighted_avg)
  Main_Matrix.append(r)

Main_Matrix
data = pd.DataFrame(Main_Matrix)
data.to_csv("drive/MyDrive/librarydata.csv")
```
