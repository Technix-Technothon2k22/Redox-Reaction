```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn import datasets

# Load data in X
X, y_true = make_blobs(n_samples=300, centers=4,
					cluster_std=0.50, random_state=0)
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

print(labels)

# Plot result

# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = ['y', 'b', 'g', 'r']
print(colors)
for k, col in zip(unique_labels, colors):
	if k == -1:
		# Black used for noise.
		col = 'k'

	class_member_mask = (labels == k)

	xy = X[class_member_mask & core_samples_mask]
	plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
			markeredgecolor='k',
			markersize=6)

	xy = X[class_member_mask & ~core_samples_mask]
	plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
			markeredgecolor='k',
			markersize=6)

plt.title('number of clusters: %d' % n_clusters_)
plt.show()

```

```python
import pandas as pd

data = pd.read_csv("drive/MyDrive/booksdata.csv", error_bad_lines=False,encoding='ISO-8859-1')
data = data.drop(['Publisher', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1)
data = data.drop([a for a in range(10000,len(data))])
data
```

```python
import math
#title matrix
Title_Matrix = []

def count_frequency(word_list):      
    D = {}
    for new_word in word_list:
        if new_word in D:
            D[new_word] = D[new_word] + 1
        else:
            D[new_word] = 1
    return D

def dotProduct(D1, D2): 
    Sum = 0.0
    for key in D1:
        if key in D2:
            Sum += (D1[key] * D2[key])
    return Sum
  
def vector_angle(D1, D2): 
    numerator = dotProduct(D1, D2)
    denominator = math.sqrt(dotProduct(D1, D1)*dotProduct(D2, D2))
    return math.acos(numerator / denominator)

def documentSimilarity(X, Y):
    X = X.lower()
    Y = Y.lower()  
    count_of_x = count_frequency(X)
    count_of_y = count_frequency(Y)
    distance = vector_angle(count_of_x, count_of_y)
    return distance      

for a in range(len(data)):
  r = []
  #find similarity of a w.r.t title with all elements in data
  t = data["Book-Title"][a]
  for b in range(a+1, len(data)):
    #check similarity of a with b
    c = documentSimilarity(t, data["Book-Title"][b])
    r.append(c)
  Title_Matrix.append(r)
```

```python
temp = []

for i in range(len(Title_Matrix)):
  tr = []
  for j in range(len(Title_Matrix[i])):
    tr.append(Title_Matrix[i][j])
  temp.append(tr)

data = pd.DataFrame(temp)
data.to_csv("drive/MyDrive/title_matrix.csv")
print("Done")
```
