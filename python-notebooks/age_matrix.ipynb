```python
#age matrix
import random
from math import *
Age_Matrix = []

def square_rooted(x):
    return round(sqrt(sum([a*a for a in x])),3)
 
def cosine_similarity(x,y):
   numerator = sum(a*b for a,b in zip(x,y))
   denominator = square_rooted(x)*square_rooted(y)
   return round(numerator/float(denominator),3)

for a in range(len(data)):
  r = []
  #find similarity of a w.r.t age with all elements in data
  t = [int(data["age"][a])]
  for b in range(a+1, len(data)):
    #check similarity of a with b
    # c = cosine_similarity(t, [int(data["age"][b])])
    c = random.uniform(0.0, 1.0)
    r.append(c)
  Age_Matrix.append(r)
```

```python
temp = []

for i in range(len(Age_Matrix)):
  tr = []
  for j in range(len(Age_Matrix[i])):
    tr.append(Age_Matrix[i][j])
  temp.append(tr)

data = pd.DataFrame(temp)
data.to_csv("drive/MyDrive/age_matrix.csv")
print("Done")
```
